```python
import pandas as pd

summary_df = pd.read_excel("~/Desktop/hla/test_cjt/data/PXD034772/DDA/dda_result_compare.xlsx")
```

```python
summary_df["pFind+whole_db"] = [0]*summary_df.shape[0]
summary_df["pFind+reduced_db"] = [0]*summary_df.shape[0]
summary_df
```

```python
import numpy as np

def cal_unique_pep_num_from_pfind_result(fname, min_pep_length=8, max_pep_length=15):
    df = pd.read_table(fname)
    unique_pep_pfind = df["Sequence"].tolist()
    unique_pep_pfind = np.unique(unique_pep_pfind)
    unique_pep_pfind = [element for element in unique_pep_pfind if ((len(element)>=min_pep_length) & (len(element)<=max_pep_length))]
    return len(unique_pep_pfind)
```

```python
base_path = "/Users/chengjintao/Desktop/hla/test_cjt/data/PXD034772/DDA"

for i in range(summary_df.shape[0]):
    tmp_id = summary_df.loc[i, "sample_id"]
    dir_id = tmp_id.split("-")[0]
    print(tmp_id)
    f1_path = f"{base_path}/{dir_id}/{tmp_id}/result/pFind-Filtered.spectra"
    f2_path = f"{base_path}/{dir_id}/{tmp_id}-0819/result/pFind-Filtered.spectra"
    summary_df.loc[i, "pFind+whole_db"] = cal_unique_pep_num_from_pfind_result(f1_path)
    summary_df.loc[i, "pFind+reduced_db"] = cal_unique_pep_num_from_pfind_result(f2_path)
summary_df
```

```python
summary_df["increase_ratio"] = [0.0]*summary_df.shape[0]

for i in range(summary_df.shape[0]):
    tmp_ratio = (summary_df.loc[i, "pFind+reduced_db"] - summary_df.loc[i, "pFind+whole_db"]) / summary_df.loc[i, "pFind+whole_db"]
    summary_df.loc[i, "increase_ratio"] = tmp_ratio
summary_df
```

```python
summary_df.to_excel("~/Desktop/hla/test_cjt/data/PXD034772/DDA/dda_result_compare.xlsx", index=False)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 8))
# 设置柱状图的位置和宽度
bar_width = 1
index = 4*np.arange(summary_df.shape[0])

# 绘制柱状图
plt.bar(index - 0.5*bar_width, summary_df["Comet+whole_db"], bar_width, label="Comet+whole_db")
plt.bar(index + 0.5*bar_width, summary_df["pFind+whole_db"], bar_width, label="pFind+whole_db")
plt.bar(index + 1.5*bar_width, summary_df["pFind+reduced_db"], bar_width, label="pFind+reduced_db")

# 添加标题和轴标签
plt.xlabel('Sample id')
plt.ylabel('Unique peptides number')

# 显示图例
plt.legend()

# 在X轴上设置类别标签
plt.xticks(index + bar_width / 2, summary_df["sample_id"], rotation=90)

# 显示图表
# plt.tight_layout()
plt.show()
```
