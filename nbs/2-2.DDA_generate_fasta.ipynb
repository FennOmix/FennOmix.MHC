```python
import pandas as pd

sample_df = pd.read_table("data/demo/peptide_df.tsv")
sample_df
```

```python
with open("data/demo/predicted.fasta", "w") as f:
    for i in range(sample_df.shape[0]):
        pep = sample_df.loc[i, "sequence"]
        f.write(f">predicted_{i}\n")
        f.write(pep)
        f.write("\n")
```
